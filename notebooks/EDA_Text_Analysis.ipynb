```python
# --- Text Analysis: Import Required Libraries ---

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
from wordcloud import WordCloud
from textblob import TextBlob
import seaborn as sns
import re
```

```python
# --- Load Raw Data ---
df = pd.read_csv("../Datas/newsData/raw_analyst_ratings.csv")

df.head()

```

```python
# --- Basic Info Check ---

df.info()
df.describe(include="all")

```

```python
# --- Clean the Headline Column ---

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'[^a-zA-Z0-9\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df["headline_clean"] = df["headline"].apply(clean_text)

df.head()

```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text_data = " ".join(df["headline"].dropna().tolist())

wordcloud = WordCloud(
    width=1200,
    height=700,
    background_color="white"
).generate(text_data)

plt.figure(figsize=(14, 8))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("WordCloud of Cleaned Headlines", fontsize=20)
plt.show()
```

```python
from collections import Counter

words = " ".join(df["headline"]).split()
word_counts = Counter(words)

top_words = word_counts.most_common(20)
top_words
```

```python
# --- Plot Top Words ---

import seaborn as sns

top_words_df = pd.DataFrame(top_words, columns=["word", "count"])

plt.figure(figsize=(12,6))
sns.barplot(data=top_words_df, x="count", y="word")
plt.title("Top 20 Most Common Words in Headlines")
plt.show()
```
